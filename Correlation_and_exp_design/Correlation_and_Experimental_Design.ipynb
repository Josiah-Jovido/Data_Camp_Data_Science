# Correlation
Correlation refers to the statistical relationship between two entities. In other words, it's how two variables move in relation to one another.
* **Positive correlation:** A positive correlation would be 1. This means the two variables moved either up or down in the   same direction together.
* **Negative correlation:** A negative correlation is -1. This means the two variables moved in opposite directions.
* **Zero or no correlation:** A correlation of zero means there is no relationship between the two variables. In other words, as one variable moves one way, the other moved in another unrelated direction.
#### The Dataset used for this analysis can be found in the repo.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Creating a scatterplot of happiness_score vs. life_exp (without a trendline) using seaborn.

```python
world_happiness = pd.read_csv('world_happiness.csv')
world_happiness.head()
```

```python
sns.scatterplot(x='life_exp', y='happiness_score', data=world_happiness)
plt.show()
```

#### Creating a scatterplot of happiness_score vs. life_exp with a linear trendline using seaborn, and setting ci to None.

```python
sns.lmplot(x='life_exp', y='happiness_score', data=world_happiness, ci=None)
plt.show()
```

#### Calculating the correlation between life_exp and happiness_score and saving it as cor.

```python
cor = world_happiness['life_exp'].corr(world_happiness['happiness_score'])
print(cor)
```

#### Creating a seaborn scatterplot (without a trendline) showing the relationship between gdp_per_cap (on the x-axis) and life_exp (on the y-axis).

```python
sns.scatterplot(x='gdp_per_cap', y='life_exp', data=world_happiness)
plt.show()
```

#### Calculating the correlation between gdp_per_cap and life_exp and storing as cor.

```python
cor = world_happiness['gdp_per_cap'].corr(world_happiness['life_exp'])
print(cor)
```

#### Creating a scatterplot of happiness_score versus gdp_per_cap and calculating the correlation between them.

```python
sns.scatterplot(x='gdp_per_cap', y='happiness_score', data=world_happiness)
plt.show()
```

```python
cor = world_happiness['gdp_per_cap'].corr(world_happiness['happiness_score'])
print(cor)
```

#### Adding a new column to world_happiness called log_gdp_per_cap that contains the log of gdp_per_cap.

```python
world_happiness['log_gdp_per_cap'] = np.log(world_happiness['gdp_per_cap'])
```

```python
sns.scatterplot(x='log_gdp_per_cap', y='happiness_score', data=world_happiness)
plt.show()
```

```python
cor = world_happiness['log_gdp_per_cap'].corr(world_happiness['happiness_score'])
print(cor)
```
